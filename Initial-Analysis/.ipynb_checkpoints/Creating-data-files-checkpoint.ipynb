```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def arr(file):
    df = pd.DataFrame()
    df['Time'] = pd.date_range('2020-12-21 23:00:00', '2023-01-10 20:00:00', freq='H')
    
    temp = pd.read_excel(file)
    temp = temp[['time', 'E_min_min']]
    temp['time'] = pd.to_datetime(temp['time'], format = '%d.%m.%Y %H:%M')
    temp.columns = ['Time', 'E_min_min']
    
    df = df.merge(temp, on='Time', how='left').fillna(np.nan)
    
    arr = [np.nan]
    i = 1
    while i <= len(df)-1:
        if df.iloc[i-1]['E_min_min']==np.nan:
            arr.append(np.nan)
        elif df.iloc[i]['E_min_min']==np.nan:
            arr.append(np.nan)
        else:
            arr.append((df.iloc[i]['E_min_min']-df.iloc[i-1]['E_min_min']))
        i+=1
    df[file] = arr
    return df[['Time',file]]
```

## Combining dataframes

```python
df = pd.DataFrame()
df['Time'] = pd.date_range('2020-12-21 23:00:00', '2023-01-10 20:00:00', freq='H')

lists = [prod_k24, prod_k28, prod_h16, prod_h22, prod_h28, prod_h32, cons_k24, cons_k28, cons_h16, cons_h22, cons_h28, cons_h32]


for l in lists:
    df = df.merge(l, on='Time')
    
df = df.set_index(df['Time']).drop(['Time'],axis=1)
df.columns = lists = ['prod_k24', 'prod_k28', 'prod_h16', 'prod_h22', 'prod_h28', 'prod_h32', 'cons_k24', 'cons_k28', 'cons_h16', 'cons_h22', 'cons_h28', 'cons_h32']
df = df.drop(df.index[0])
df.to_csv('complete_dataset.csv')
```
