```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams["figure.figsize"] = (20,3)
```

```python
el = pd.read_csv('Elspotprices.csv', sep=';', decimal=',')
```

```python
el = el.drop(['HourUTC','SpotPriceEUR'],axis=1)
el['HourDK'] = pd.to_datetime(el['HourDK'])
el = el.set_index('HourDK')
```

```python
plt.plot(el[el['PriceArea']=='DK1']['SpotPriceDKK'],label='DK1',alpha=0.5)
plt.plot(el[el['PriceArea']=='DK2']['SpotPriceDKK'],label='DK2',alpha=0.5)
plt.legend()
plt.show()
```

```python
el['Year'] = el.index.year
el['Month'] = el.index.month
el['Day'] = el.index.day
el['Hour'] = el.index.hour
```

```python
elDK1 = el[el['PriceArea']=='DK1']
elDK2 = el[el['PriceArea']=='DK2']
```

```python
low_price = elDK1[(elDK1['Year']>=2018) & (elDK1['Year']<2022)]
```

```python
high_price = elDK1[(elDK1['Year']>=2022)]
```

```python
low_price.groupby(['Month','Day','Hour'],as_index=False).mean()['SpotPriceDKK'].plot()
high_price.groupby(['Month','Day','Hour'],as_index=False).mean()['SpotPriceDKK'].plot()
plt.show()
```

## DK1 data from 2013

```python
df = pd.read_csv('ElspotpricesDK1.csv', sep=';', decimal=',')
df = df[['HourDK','SpotPriceDKK']]
df.loc[df['SpotPriceDKK']>10000,'SpotPriceDKK'] = df[df['HourDK']=='2013-06-07 06:00']
df['HourDK'] = pd.to_datetime(df['HourDK'])
```

```python
fi = pd.DataFrame()
fi['HourDK'] = pd.date_range(df['HourDK'].min(),df['HourDK'].max(),freq='H')
```

```python
fi = df.merge(fi, on='HourDK', how='right').fillna(method="ffill")
```

```python
fi = fi.set_index('HourDK')
fi.plot()
```

```python
fi['Year'] = fi.index.year
fi['Month'] = fi.index.month
fi['Day'] = fi.index.day
fi['Hour'] = fi.index.hour
```

```python
 fi.groupby(['Month','Day','Hour'],as_index=False).mean()
```

```python
all_price = fi.groupby(['Month','Day','Hour'],as_index=False).mean()
high_price = fi[((fi['Year']==2021)&(fi['Month']>=6))^(fi['Year']>2021)].groupby(['Month','Day','Hour'],as_index=False).mean()
low_price = fi[((fi['Year']==2021)&(fi['Month']<6))^(fi['Year']<2021)].groupby(['Month','Day','Hour'],as_index=False).mean()

all_price = all_price.drop(all_price[((all_price['Month']==2)&(all_price['Day']==29))].index)
low_price = low_price.drop(low_price[((low_price['Month']==2)&(low_price['Day']==29))].index)
```

```python
dR_noLeap = pd.date_range('2021-01-01 00:00:00','2021-12-31 23:00:00', freq='H')

all_price = all_price.set_index(dR_noLeap).drop(['Year','Month','Day','Hour'],axis=1)
high_price = high_price.set_index(dR_noLeap).drop(['Year','Month','Day','Hour'],axis=1)
low_price = low_price.set_index(dR_noLeap).drop(['Year','Month','Day','Hour'],axis=1)
```

```python
high_price['SpotPriceDKK'].plot(label='high')
low_price['SpotPriceDKK'].plot(label='low')
all_price['SpotPriceDKK'].plot(label='all')
plt.legend()
plt.show()
```

```python
high_price['SpotPriceDKK'].rolling(20,center=True).mean().plot(label='high')
low_price['SpotPriceDKK'].rolling(20,center=True).mean().plot(label='low')
all_price['SpotPriceDKK'].rolling(20,center=True).mean().plot(label='all')
plt.legend()
plt.show()
```
