```python
def plott(filename):
    df = pd.read_csv(filename)
    #print(df.groupby(['sample_size']).mean())
    
    fig,axs = plt.subplots(2,1,figsize=(15,6),sharex=True)
    sns.boxplot(ax=axs[0], data=df, x="sample_size", y="rmse")
    sns.boxplot(ax=axs[1], data=df, x="sample_size", y="p_value")
    plt.show()
```

```python
plott(prod)
plott(cons)
```

```python
def ttest(filename):
    df = pd.read_csv(filename)
    arrs = pd.DataFrame()
    for i in df['sample_size'].unique():
        col_name = 'ss_'+str(i)
        arrs[col_name] = df[df['sample_size']==i]['rmse'].to_numpy()
        
    num_cols = arrs.shape[1]
    for i in range(num_cols):
        for j in range(i+1,num_cols):
            print(arrs.iloc[:,i].name, arrs.iloc[:,j].name)
            print(np.round(stats.ttest_ind(a=arrs.iloc[:,i], b=arrs.iloc[:,j], equal_var=True)[1],4))  
            print('')
```

```python
ttest(cons)
```
